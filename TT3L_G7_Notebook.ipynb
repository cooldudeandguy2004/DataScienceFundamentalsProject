**Brief Description:**  
This dataset includes global sustainable energy indicators for the years 2000 - 2020 for countries around the world. It comprises of variables like access to electricity, access to clean fuels for cooking, share of renewable energy, share of electricity generation from fossil fuels and renewables, share of low-carbon electricity generation, energy intensity, GDP per capita and CO₂ emissions. The dataset can be used for the TT3L domain, Environment, Climate & Sustainability, as it can be analysed in relation with sustainable energy transition, energy efficiency and energy-related emissions.

**Problem Statement:**  
Access to reliable energy and reduction in fossil fuel consumption and carbon emissions are areas of focus that many countries are attempting to improve upon. However, there are variations in progress between countries in terms of energy infrastructure, use of renewables, economic development and energy efficiency. This project aims to use global sustainable energy data between 2000 and 2020 to detect trends, relationships, predicting factors and country clusters that can help to improve the sustainability and climate-related decision-making process.

**Analytical Questions:**  
Q1. How did global sustainable energy indicators change from 2000 to 2020?

Q2. Are countries with different GDP per capita levels significantly different in clean energy access and energy intensity?

Q3. Which energy and economic indicators can predict CO₂ emissions?

Q4. Can countries be grouped based on their sustainable energy profiles?
