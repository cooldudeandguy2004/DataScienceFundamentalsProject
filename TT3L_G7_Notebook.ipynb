## 4. Data Cleaning & Preprocessing
- Handle missing values (strategy and justification)
- Fix types (dates, categoricals, numerics)
- Remove/flag outliers
- Feature engineering (document purpose)
- Keep transformations transparent and reproducible


```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_', regex=False)
    .str.replace('/', '_', regex=False)
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.replace('%', 'pct', regex=False)
    .str.replace('$', 'usd', regex=False)
    .str.replace(r'\\n', '_', regex=True)
)

df['density_p_km2'] = df['density_p_km2'].astype(str).str.replace(',', '', regex=False).astype(float)

print("Missing values before cleaning:\n", df.isnull().sum())

missing_threshold = len(df) * 0.5
df = df.dropna(thresh=missing_threshold, axis=1)

import numpy as np
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df.groupby('entity')[col].transform('median'))
        df[col] = df[col].fillna(df[col].median())

print("\nMissing values after cleaning:\n", df.isnull().sum())

print("\nDuplicates before cleaning:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after cleaning:", df.duplicated().sum())
```

### Data Cleaning Justification



**1. Standardizing Column Names:**
All column names were converted to lowercase, whitespaces stripped and replaced with underscores and special characters. It eliminates syntax errors, and makes it easier to access columns by either the . notation or by a dictionary key in the analysis.

**2. Fixing Data Types:**
The population density column (`density_p_km2`) was imported as an `object` data type because the thousands separator used in the data was a comma (e.g., `1,234`). We stripped out the commas and cast the column to `float` so that we could perform numerical operations.

**3. Handling Missing Values:**
- **Column Dropping:** We excluded any columns having more than half of the data missing (Column Dropping). Then there would be too much noise, and it would lead us astray.
- **Imputation:** For the missing numerical values, we attempted to impute using the median of that particular country to ensure geographic and economic consistency. If there was no data for that column at all, then we used the global median to keep the row being used, but not altering the global distribution significantly.

**4. Duplicates:**
We performed a duplicate check and eliminated any duplicate rows to avoid double counting.

### Q2. Are countries with different GDP per capita levels significantly different in clean energy access and energy intensity?

This question explores the extent to which there are differences in sustainable energy indicators between countries of different economic development. The records are divided into four economic groups: low GDP, lower-middle GDP, upper-middle GDP, and high GDP by dividing GDP per capita.

The analysis is based on three indicators:

- Access to electricity
- Access to clean fuels for cooking
- Energy intensity

The Kruskal-Wallis test is used as it compares more than two groups and does not assume a normal distribution.

```python
from scipy.stats import kruskal

source_df = df.copy()

q2_df = source_df[
    [
        "entity",
        "year",
        "gdp_per_capita",
        "access_to_electricity_pct_of_population",
        "access_to_clean_fuels_for_cooking",
        "energy_intensity_level_of_primary_energy_mj_usd2017_ppp_gdp"
    ]
].copy()

q2_df = q2_df.rename(columns={
    "entity": "country",
    "year": "year",
    "gdp_per_capita": "gdp_per_capita",
    "access_to_electricity_pct_of_population": "access_electricity_pct",
    "access_to_clean_fuels_for_cooking": "access_clean_cooking_pct",
    "energy_intensity_level_of_primary_energy_mj_usd2017_ppp_gdp": "energy_intensity"
})

numeric_cols_q2 = [
    "gdp_per_capita",
    "access_electricity_pct",
    "access_clean_cooking_pct",
    "energy_intensity"
]

for col in numeric_cols_q2:
    q2_df[col] = pd.to_numeric(q2_df[col], errors="coerce")

q2_df = q2_df.dropna(subset=numeric_cols_q2)

print("Q2 dataset shape after removing missing values:", q2_df.shape)

q2_df["gdp_group"] = pd.qcut(
    q2_df["gdp_per_capita"].rank(method="first"),
    q=4,
    labels=["Low GDP", "Lower-middle GDP", "Upper-middle GDP", "High GDP"]
)

print("\nNumber of records in each GDP group:")
display(q2_df["gdp_group"].value_counts().sort_index())

q2_summary = q2_df.groupby("gdp_group", observed=True)[
    ["access_electricity_pct", "access_clean_cooking_pct", "energy_intensity"]
].mean().round(2)

print("\nAverage sustainable energy indicators by GDP group:")
display(q2_summary)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=q2_df, x="gdp_group", y="access_electricity_pct")
plt.title("Access to Electricity by GDP Per Capita Group")
plt.xlabel("GDP Per Capita Group")
plt.ylabel("Access to Electricity (% of Population)")
plt.xticks(rotation=20)
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=q2_df, x="gdp_group", y="access_clean_cooking_pct")
plt.title("Access to Clean Fuels for Cooking by GDP Per Capita Group")
plt.xlabel("GDP Per Capita Group")
plt.ylabel("Access to Clean Fuels for Cooking (%)")
plt.xticks(rotation=20)
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=q2_df, x="gdp_group", y="energy_intensity")
plt.title("Energy Intensity by GDP Per Capita Group")
plt.xlabel("GDP Per Capita Group")
plt.ylabel("Energy Intensity")
plt.xticks(rotation=20)
plt.show()
```

```python
q2_test_results = []

test_variables = {
    "Access to Electricity": "access_electricity_pct",
    "Access to Clean Fuels for Cooking": "access_clean_cooking_pct",
    "Energy Intensity": "energy_intensity"
}

for variable_name, variable_col in test_variables.items():
    groups = [
        group[variable_col].dropna().values
        for _, group in q2_df.groupby("gdp_group", observed=True)
    ]
    
    stat, p_value = kruskal(*groups)
    
    q2_test_results.append({
        "Variable": variable_name,
        "Test": "Kruskal-Wallis",
        "Statistic": round(stat, 4),
        "p-value": round(p_value, 6),
        "Significant Difference?": "Yes" if p_value < 0.05 else "No"
    })

q2_test_results_df = pd.DataFrame(q2_test_results)

print("Kruskal-Wallis test results:")
display(q2_test_results_df)
```

This analysis contrasts indicators of sustainable energy within four groups of GDP per capita: Low GDP, Lower-middle GDP, Upper-middle GDP and High GDP.

Overall, the boxplots indicate that countries with higher GDPs have, on average, greater access to electricity and clean fuels for cooking than countries with lower GDPs. This implies that enhancing economic development is linked to improving energy access.

Whether the difference among the GDP groups is statistically significant was checked using the Kruskal-Wallis test. p-value < 0.05 is considered statistically significant.

There are significant differences between the access to electricity, energy intensity and access to clean fuels for cooking among the GDP per capita groups based on the test results. This indicates that the level of energy access and energy efficiency are linked to GDP per capita.

### Q3. Which energy and economic indicators can predict CO₂ emissions?

This analysis uses a supervised machine learning regression model to predict CO₂ emissions using selected energy and economic indicators. The variable of interest is value_co2_emissions_kt_by_country.

The non-linear relationship of features and the ability to display feature importance is the reason for using a Random Forest Regressor. Feature importance is useful for pinpointing the variables that are most useful for predicting CO₂ emissions.

This analysis helps to answer the project problem statement by pinpointing key energy and economic considerations related to climate-related emissions.

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

model_df_source = df.copy()

if "RANDOM_STATE" not in globals():
    RANDOM_STATE = 42

target_col = "value_co2_emissions_kt_by_country"

selected_features = [
    "year",
    "access_to_electricity_pct_of_population",
    "access_to_clean_fuels_for_cooking",
    "renewable-electricity-generating-capacity-per-capita",
    "renewable_energy_share_in_the_total_final_energy_consumption_pct",
    "electricity_from_fossil_fuels_twh",
    "electricity_from_nuclear_twh",
    "electricity_from_renewables_twh",
    "low-carbon_electricity_pct_electricity",
    "primary_energy_consumption_per_capita_kwh_person",
    "energy_intensity_level_of_primary_energy_mj_usd2017_ppp_gdp",
    "gdp_growth",
    "gdp_per_capita",
    "density_p_km2",
    "land_areakm2"
]

q3_required_cols = selected_features + [target_col]
missing_cols = [col for col in q3_required_cols if col not in model_df_source.columns]

if len(missing_cols) > 0:
    print("Missing columns:")
    for col in missing_cols:
        print("-", col)
    raise KeyError("Some required Q3 columns are missing. Check column names.")
else:
    print("All required Q3 columns are available.")

q3_df = model_df_source[q3_required_cols].copy()

for col in q3_required_cols:
    q3_df[col] = (
        q3_df[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .replace("nan", np.nan)
    )
    q3_df[col] = pd.to_numeric(q3_df[col], errors="coerce")

q3_df = q3_df.dropna(subset=[target_col])

q3_df = q3_df[q3_df[target_col] >= 0]

print("Q3 dataset shape after preparation:", q3_df.shape)
print("Number of predictor variables:", len(selected_features))

display(q3_df.head())
```

```python
X = q3_df[selected_features]
y_original = q3_df[target_col]

y = np.log1p(y_original)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE
)

rf_model = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("model", RandomForestRegressor(
        n_estimators=200,
        random_state=RANDOM_STATE
    ))
])

rf_model.fit(X_train, y_train)

print("Random Forest regression model trained successfully.")
print("Training records:", X_train.shape[0])
print("Testing records:", X_test.shape[0])
```

```python
y_pred_log = rf_model.predict(X_test)

y_test_original = np.expm1(y_test)
y_pred_original = np.expm1(y_pred_log)

mae = mean_absolute_error(y_test_original, y_pred_original)
rmse = np.sqrt(mean_squared_error(y_test_original, y_pred_original))
r2_log = r2_score(y_test, y_pred_log)
r2_original = r2_score(y_test_original, y_pred_original)

model_results = pd.DataFrame({
    "Metric": [
        "Mean Absolute Error",
        "Root Mean Squared Error",
        "R² Score on Log Scale",
        "R² Score on Original Scale"
    ],
    "Value": [
        round(mae, 2),
        round(rmse, 2),
        round(r2_log, 4),
        round(r2_original, 4)
    ]
})

print("Model evaluation results:")
display(model_results)
```

### Q4: Are countries able to be clustered by their sustainable energy profiles?


```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# Copy the dataframe
clustering_df = df.copy()

if "RANDOM_STATE" not in globals():
    RANDOM_STATE = 42

# Define features that describe the sustainable energy profile of a country
profile_features = [
    "access_to_electricity_pct_of_population",
    "access_to_clean_fuels_for_cooking",
    "renewable_energy_share_in_the_total_final_energy_consumption_pct",
    "low-carbon_electricity_pct_electricity",
    "energy_intensity_level_of_primary_energy_mj_usd2017_ppp_gdp",
    "gdp_per_capita"
]

# Aggregate features per country by taking the mean over the 2000-2020 period
country_profiles = clustering_df.groupby("entity")[profile_features].mean().dropna()

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(country_profiles)

# Determine optimal clusters using Elbow Method and Silhouette Score
wcss = []
silhouette_scores = []
k_range = range(2, 9)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))

# Plot Elbow and Silhouette curves
fig, ax1 = plt.subplots(figsize=(10, 5))

color = "tab:red"
ax1.set_xlabel("Number of Clusters (k)")
ax1.set_ylabel("WCSS (Inertia)", color=color)
ax1.plot(k_range, wcss, color=color, marker="o", label="WCSS")
ax1.tick_params(axis="y", labelcolor=color)

ax2 = ax1.twinx()
color = "tab:blue"
ax2.set_ylabel("Silhouette Score", color=color)
ax2.plot(k_range, silhouette_scores, color=color, marker="s", label="Silhouette Score")
ax2.tick_params(axis="y", labelcolor=color)

plt.title("Elbow Method & Silhouette Scores for Optimal k in Country Clustering")
fig.tight_layout()
plt.show()
```

```python
# From the plots, k=3 is the optimal choice (clear elbow and high silhouette score)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=RANDOM_STATE, n_init=10)
country_profiles["cluster"] = kmeans.fit_predict(scaled_features)

# Compute cluster profiles
cluster_profiles = country_profiles.groupby("cluster").mean().round(2)
cluster_profiles["country_count"] = country_profiles.groupby("cluster").size()

print("Cluster Profiles:")
display(cluster_profiles)
```

```python
from sklearn.decomposition import PCA

# Visualize clusters using PCA
pca = PCA(n_components=2, random_state=RANDOM_STATE)
pca_coords = pca.fit_transform(scaled_features)
country_profiles["pca1"] = pca_coords[:, 0]
country_profiles["pca2"] = pca_coords[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="pca1", y="pca2", hue="cluster", palette="Set1",
    data=country_profiles, s=100, alpha=0.8
)
plt.title("Country Clusters Visualized via PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# List countries in each cluster
for c in range(optimal_k):
    cluster_countries = country_profiles[country_profiles["cluster"] == c].index.tolist()
    print(f"\nCluster {c} Countries (first 15):")
    print(", ".join(cluster_countries[:15]) + ("..." if len(cluster_countries) > 15 else ""))
```

**Interpretations:**

#### Limitations and possible biases
* **Imputation Bias:** Imputing missing values using country-level median group-by methods, while practical, might smooth out local variations and mask sudden policy or infrastructure shifts.
* **Aggregation Levels:** The data is aggregated at the national level, which hides regional disparities within countries (e.g., urban vs. rural electrification gaps).
* **Temporal Lag:** The dataset ends in 2020. Major post-pandemic energy shifts, geopolitical impacts on fossil fuel supply chains, and recent renewable investments are not captured.

#### Logical next steps or improvements
* **Incorporate Policy & Climate Indexes:** Integrate policy indexes (like carbon taxes or subsidies) or geographical climate risk indexes to measure policy effectiveness.
* **Time-Series Forecasting:** Implement forecasting models (e.g., Prophet or LSTM) to project emissions trends up to 2030 (the target year for UN Sustainable Development Goals).
