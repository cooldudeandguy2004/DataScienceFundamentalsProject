```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- TEMPORARY DATA LOADING FOR MEMBER 2 ---
# This will be replaced by Member 1's code later.
df = pd.read_csv('global-data-on-sustainable-energy.csv.xls')

# Check basic info
print(df.info())
df.head()
```

## 4. Data Cleaning & Preprocessing
- Handle missing values (strategy and justification)
- Fix types (dates, categoricals, numerics)
- Remove/flag outliers
- Feature engineering (document purpose)
- Keep transformations transparent and reproducible


```python
# 1. Standardize Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('%', 'pct').str.replace('$', 'usd', regex=False).str.replace('\n', '', regex=False)

# 2. Fix Data Types
# 'density_p/km2' was loaded as an object due to commas (e.g. '1,234'). Convert to float.
df['density_p/km2'] = df['density_p/km2'].astype(str).str.replace(',', '', regex=False).astype(float)

# 3. Handle Missing Values
print("Missing values before cleaning:\n", df.isnull().sum())

# Drop columns with excessive missing values (>50%)
missing_threshold = len(df) * 0.5
df = df.dropna(thresh=missing_threshold, axis=1)

# Impute remaining numerical columns
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        # Impute with country median first
        df[col] = df[col].fillna(df.groupby('entity')[col].transform('median'))
        # If a whole country is NaN, fallback to global median
        df[col] = df[col].fillna(df[col].median())

print("\nMissing values after cleaning:\n", df.isnull().sum())

# 4. Check for duplicates
print("\nDuplicates before cleaning:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after cleaning:", df.duplicated().sum())
```

### Q1: Trend Analysis (Member 2)
**Question:** How has the global share of renewable energy consumption and electricity generated from renewables trended from 2000 to 2020?


```python
# Group by Year to see global trends
global_trend = df.groupby('year')[['renewable_energy_share_in_the_total_final_energy_consumption_pct', 'electricity_from_renewables_twh']].mean().reset_index()

plt.figure(figsize=(12, 5))

# Plot 1: Renewable Energy Share
plt.subplot(1, 2, 1)
sns.lineplot(data=global_trend, x='year', y='renewable_energy_share_in_the_total_final_energy_consumption_pct', marker='o')
plt.title('Global Trend: Renewable Energy Share (%)')
plt.ylabel('Share of Total Consumption (%)')

# Plot 2: Electricity from Renewables
plt.subplot(1, 2, 2)
sns.lineplot(data=global_trend, x='year', y='electricity_from_renewables_twh', marker='o', color='green')
plt.title('Global Trend: Electricity from Renewables (TWh)')
plt.ylabel('TWh')

plt.tight_layout()
plt.show()
```
